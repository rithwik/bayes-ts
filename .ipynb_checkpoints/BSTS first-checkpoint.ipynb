Based on the bstspost.Rmd post.

Getting predictions that make sense seems to be non-trivial compared to the usual methods of running a 'predict' method on the test set. In fact the below results are on the train set itself as I am yet to figure out how it applies to the test set.

For the train set the fit seems quite good.

```r
library(data.table)
library(dplyr)
library(bsts)

raw_df <- fread('/home/rithwik/egovs/data/cocCombined.csv')
raw_df$Complaint.Date <- as.Date(raw_df$Complaint.Date, format = "%m/%d/%Y")
raw_df$Resolution.Date <- as.Date(raw_df$Resolution.Date, format = "%m/%d/%Y")
raw_df$X <- NULL
raw_df$V1 <- NULL

df <- raw_df %>% 
  filter(Complaint.Type == 'Non Burning of Street Lights') %>% 
  group_by(Complaint.Date) %>% 
  summarize(count = n())

dates <- data.frame(
  Complaint.Date = seq(min(df$Complaint.Date), max(df$Complaint.Date), 1)
)

df <- left_join(dates, df)
df[is.na(df$count), 'count'] <- 0

train <- df %>% 
  filter(#Complaint.Date >= as.Date('2012-01-01', '%Y-%m-%d') &
           Complaint.Date <= as.Date('2014-12-31', '%Y-%m-%d'))
  
test <- df %>% 
  filter(Complaint.Date >= as.Date('2015-01-01', '%Y-%m-%d') &
           Complaint.Date <= as.Date('2015-12-31', '%Y-%m-%d'))
```

```r
library(lubridate)
library(ggplot2)

y <- train$count # log10(train)
ss <- AddLocalLinearTrend(list(), y)
ss <- AddSeasonal(ss, y, nseasons = 12)
bsts.model <- bsts(y, state.specification = ss, niter = 500, 
                   ping = 0, seed = 2016)
```

```r
### Get a suggested number of burn-ins
burn <- SuggestBurn(0.1, bsts.model)

### Predict
p <- predict.bsts(bsts.model, horizon = 12, burn = burn, quantiles = c(.025, .975))

### Actual versus predicted
d2 <- data.frame(
  # fitted values and predictions
  c(as.numeric(-colMeans(bsts.model$one.step.prediction.errors[-(1:burn),])+y)),
  # actual data and dates 
  train$count,
  train$Complaint.Date)
names(d2) <- c("Fitted", "Actual", "Date")

### MAPE
MAPE <- filter(d2, year(Date)>=2015) %>% 
  summarise(MAPE=mean(abs(Actual-Fitted)/Actual))
```

```r
### Plot
ggplot(data=d2, aes(x=Date)) +
  geom_line(aes(y=Actual, colour = "Actual"), size=1.2) +
  geom_line(aes(y=Fitted, colour = "Fitted"), size=1.2, linetype=2) +
  theme_bw() + theme(legend.title = element_blank()) + ylab("") + xlab("") +
  ggtitle(paste0("BSTS -- Holdout MAPE = ", round(100*MAPE,2), "%")) +
  theme(axis.text.x=element_text(angle = -90, hjust = 0))
```
